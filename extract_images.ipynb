```python
# Importing all necessary libraries 
import cv2 
import os 
from time import time
  
# Read the video from specified path 
VIDEO_NAMES = ['2020-07-01_Egypt_1',              
                '2020-07-01_Harvest_1',
                '2020-07-01_Harvest_2',
                '2020-07-01_Harvest_3',
                '2020-07-01_Harvest_4',
                '2020-07-01_Harvest_5',
                '2020-07-01_Harvest_6']

for file_name in VIDEO_NAMES:
    new_folder_path = './data/' + file_name
    video_path = "C:\\Users\\Edward\\Videos\\Team Fortress 2\\" + file_name + ".mp4"
    cam = cv2.VideoCapture(video_path) 
    
    # frame 
    currentframe = 0

    # only choose every 5 frames
    frame_skip = 75

    start = time()

    while(True): 

        # reading from frame 
        ret, frame = cam.read() 

        if ret: 
            if currentframe % frame_skip == 0:
                # if video is still left continue creating images 
                name = os.path.join(os.curdir, 'data', 'frames', file_name + "_frame" + str(currentframe) + '.jpg')
                if currentframe % (4 * frame_skip) == 0:
                    print('Creating...' + name) 
        
                # writing the extracted images 
                if not cv2.imwrite(name, frame):
                    raise Exception("Could not write image to {}".format(name))

            # increasing counter so that it will 
            # show how many frames are created 
            currentframe += 1
        else: 
            break

    stop = time()
    print('DONE: {} \n\tTime taken: {}\n'.format(file_name, stop-start))

    # Release all space and windows once done 
    cam.release() 
    cv2.destroyAllWindows() 
```
